```python
# Import Packages
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import ppscore as pps
import missingno
import os
```

# Methods

## 1. Column Nomenclature

```python
def standardize_column_names(df):
    df.columns = [i.replace(' ', '_').lower() for i in df.columns]
    return df
```

## 2. Understanding our Data

```python
def eda(df):
    dic = {}
    print('Calculating dtypes...')
    dic['dtypes'] = df.dtypes
    print('Calculating count...')
    dic['count'] = df.count()
    print('Calculating null_sum...')
    dic['null_sum'] = df.isnull().sum()
    print('Calculating null_pct...')
    dic['null_pct'] = df.isnull().mean()
    print('Calculating nunique...')
    dic['nunique'] = df.nunique()
    print('Calculating min...')
    dic['min'] = df.min()
    print('Calculating 25%...')
    dic['25%'] = df.quantile(0.25)
    print('Calculating 50%...')
    dic['50%'] = df.quantile(0.5)
    print('Calculating 75%...')
    dic['75%'] = df.quantile(0.75)
    print('Calculating max...')
    dic['max'] = df.max()
    print('Calculating mean...')
    dic['mean'] = df._get_numeric_data().mean()
    print('Calculating median...')
    dic['median'] = df._get_numeric_data().median()
    print('Calculating std...')
    dic['std'] = df._get_numeric_data().std()
    print('Calculating skew...')
    dic['skew'] = df._get_numeric_data().skew()
    eda = pd.DataFrame(dic)
    eda = eda.fillna('-').round(3)
    return eda
```

## 3. Dropping Features

```python
def drop_columns(df, drop_columns):
    df.drop(drop_columns, inplace=True, axis=1, errors='ignore')
    return df
```

## 4. Dealing with Feature Data Types

```python
def convert_columns_to_date(df, columns):
    for column_name in columns:
        df[column_name] = df[column_name].apply(pd.to_datetime, infer_datetime_format=True, errors='coerce')
    return df

def convert_columns_to_category(df, columns, ordered):
    for i,column_name in enumerate(columns):
        df[column_name] = pd.Categorical(df[column_name], ordered=ordered[i])
    return df
```

## 5. Handling Missing/Duplicate Values

```python
def visualize_nulls(df):
    missingno.matrix(df)
    plt.show()
    
def drop_na_rows(df):
    df.dropna(inplace=True)
    return df

def fill_na_rows(df):
    df.interpolate(method='pad', limit_direction='forward', inplace=True)
    return df

def drop_duplicates_rows(df):
    df.drop_duplicates(inplace=True)
    return df
```

## 6. Exploring Numerical Features

```python
def find_correlation(df):
    corr = df.corr().round(2)
    return corr.style.background_gradient(cmap='coolwarm', axis=None).set_precision(2)

def boxplot_of_numerical_columns(df_in):
    sns.set(style="darkgrid")
    columns = df_in.select_dtypes(include=np.number).columns
    figure = plt.figure(figsize=(20, 30))
    figure.add_subplot(1, len(columns), 1)
    for index, col in enumerate(columns):
        if index > 0:
            figure.add_subplot(1, len(columns), index + 1)
            sns.boxplot(x=col, data=df, boxprops={'facecolor': 'None'})
            plt.show()
```

## 7. Exploring Categorical Features

```python
def plot_categorical_column(df_in, column_name, hue=None):
    sns.set(style="darkgrid")
    plot_order = df_in[column_name].value_counts(dropna=False).index
    fig = sns.catplot(y=column_name, kind="count", data=df_in, hue=hue, order=plot_order, aspect=3)
    plt.show()

def get_summary_of_categorical_column(df_in, column_name, hue=None):
    dic = {}
    dic['count'] = df_in[column_name].value_counts(dropna=False)
    dic['relative_pct'] = df_in[column_name].value_counts(dropna=False, normalize=True)
    df_out = pd.DataFrame(dic)
    return df_out
```

## 8. Plotting Numerical versus Categorical Features

```python
def plot_numerical_versus_categorical(df, numeric_column, categorical_column):
    fig = sns.catplot(x=categorical_column, y=numeric_column, kind='violin', data=df, aspect=3)
    fig.set_xticklabels(rotation=90)
    plt.show()
    
def find_predictive_power_score(df):
    pscore = pps.matrix(df).round(2)
    return pscore.style.background_gradient(cmap='coolwarm', axis=None).set_precision(2)
```

## 9. Handling Data Outliers

```python
# Incase a column has too many outliers which can affect analysis, they can be removed with this
def remove_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = df_in.loc[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    return df_out
```
